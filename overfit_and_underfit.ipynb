# 映画レビュー分類モデル

```
import keras import model
import keras import layers

model = models.Sequential()
model.add(models.Dense(16, activation='relu', input_shape=(10000, )))
model.add(models.Dense(16, activation='relu'))
model.add(models.Dense(1, activation='sigmoid'))
```

**キャパシティが小さいモデル**

```
model = models.Sequential()
model.add(models.Dense(4, activation='relu', input_shape=(10000, )))
model.add(models.Dense(4, activation='relu'))
model.add(models.Dense(1, activation='sigmoid'))
```

**キャパシティが大きいモデル**

```
model = model.Sequential()
model.add(layers.Dense(512, activation='relu', input_shape=(10000, )))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

**モデルにL2正則化を追加**

```
from keras import regularizers

model = models.Sequential()
model.add(layers.Dense(16, kernel_regularizer=regularizers.12(0.001), activation='relu', input_shape=(10000,)))
model.add(layers.Dense(16, kernel_regularizer=regularizers,12(0.001), activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

**重みを正則化するための正則化項**

```
from keras import regularizers

#L1正則化
regularizers.l1(0.001)
#L1正則化とL2正則化の同時適用
regularizers.l1_l2(l1=0.001, l2=0.001)
```

**ドロップアウトを追加**

```
model = models.Sequential()
model.add(layers.Dense(16, activation='relu', input_shape=(10000, )))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))
```
